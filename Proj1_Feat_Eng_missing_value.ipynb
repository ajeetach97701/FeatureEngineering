```python
from sklearn.impute import SimpleImputer
impute = SimpleImputer(strategy='most_frequent')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
```

```python
X_train
```

```python
# Missing value imputer
transformer = ColumnTransformer(
        transformers=[
            ('tnf1', SimpleImputer(strategy='most_frequent'), ['Embarked']),
            ('tnf2', SimpleImputer(), ['Age']),
            ('tnf3', SimpleImputer(strategy='constant'), ['Cabin']),
        ],
        verbose_feature_names_out= False,
        remainder='passthrough',
    )

#
```

```python
X_train = transformer.fit_transform(X_train)
X_train = pd.DataFrame(X_train, columns=transformer.get_feature_names_out())
```

```python
X_test = transformer.fit_transform(X_test)
X_test = pd.DataFrame(X_test, columns=transformer.get_feature_names_out())
```

```python
# One Hot Encoding on Sex and Embarked columns
processor = ColumnTransformer(transformers=[
               ('encSex', OneHotEncoder(sparse=False, dtype=np.int32, drop='first'), ['Sex']),
            ('encEmbarked', OneHotEncoder(sparse=False, dtype=np.int32, drop='first'), ['Embarked']),
  
],verbose_feature_names_out=False
                              ,remainder='passthrough')
        
```

```python
X_train = processor.fit_transform(X_train)
X_train = pd.DataFrame(X_train, columns=processor.get_feature_names_out())
X_train
```

```python
X_test = processor.fit_transform(X_test)
X_test = pd.DataFrame(X_test, columns=processor.get_feature_names_out())
```
